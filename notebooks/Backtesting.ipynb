```python
def calculate_returns(self):
    return (self.wallet_worth_history[-1] - self.wallet_worth_history[0]) / self.wallet_worth_history[0]
```

```python
def calculate_max_drawdown(self):
    return (min(self.wallet_worth_history) - max(self.wallet_worth_history)) / max(self.wallet_worth_history)
```

```python
def pct_change(x):
    pct_change = []
    for i in range(1, len(x)):
        change = round((x[i] - x[i-1]) / x[i-1], 6)
        pct_change.append(change)
    return np.array(pct_change)
```

```python
returns = pct_change(portfolio.wallet_worth_history)
```

```python
plt.plot(portfolio.wallet_worth_history)
```

```python
plt.plot(100*returns.cumsum());
plt.ylabel('Total Returns (%)')
plt.title('Buying/Selling at random');
```

```python
def sharpe_ratio(returns, N=60*24*365, risk_free_rate=0):
    mean = returns.mean() * N - risk_free_rate
    sigma = returns.std() * np.sqrt(N)
    return mean / sigma
```

```python
def sortino_ratio(returns, N=60*24*365, risk_free_rate=0):
    mean = returns.mean() * N - risk_free_rate
    negative_std = returns[returns < 0].std() * np.sqrt(N)
    return mean / negative_std
```

```python
plt.hist(100*returns, bins=50);
```

```python
# Sharpe
sharpe_ratio(returns, N)
```

```python
# Sortino
sortino_ratio(returns, N)
```
